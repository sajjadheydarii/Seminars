# **Visualization**

```python
x = np.linspace(0,2,100)

plt.plot(x, x, label = "linear", color = "lime", linestyle = "--")
plt.plot(x, x** 2, label = "quadratic", color = "tomato", linestyle = "-.", linewidth = 7)
plt.plot(x, x** 3, label = "cubic", color = "darkviolet", linestyle = "dotted", linewidth = 5)

plt.title("Simple Plots")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.legend(loc = "lower right")
plt.show()
```

```python
years = [1960,1970,1980,1990,2000,2010]
Iran_pop = [21.91,28.51,38.94,56.54,66.16,75.24]
Iraq_pop = [27.47,34.88,43.98,53.92,63.24,71.25]

plt.plot(years, Iran_pop, linestyle = "-", marker = "s", mew = 5)
plt.plot(years, Iraq_pop, linestyle = "--", color = "red")

plt.title("Iran vs Iraq population Growth Rate")
plt.legend(["Iran", "Iraq"])

plt.xlabel("Year")
plt.ylabel("Population")

plt.yticks([20,30,40,50,60,70,80],
           ["20million","30million","40million","50million","60million","70million","80million"])

plt.annotate("Iran-Iraq War",
             xytext = (1996,35),
             fontsize = 12,
             xy = (1980,39),
             arrowprops = dict(facecolor = "silver", width = 2))
plt.grid()
plt.show()
```

```python
data = {"Iran" : 80, "Italy" : 53, "Sweden" : 20, "Turkey" : 68}

names = list(data.keys())
values = list(data.values())

plt.barh(names,values)

plt.title("Country Population")
plt.xlabel("Population")
plt.ylabel("Country Name")

plt.show()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression        #chi2

# generate dataset
X = tips.iloc[:,[1,6]]
Y = tips.iloc[:,0]

# define feature selection
fs = SelectKBest(score_func=f_regression, k=2)
fit = fs.fit(X, Y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

# apply feature selection
featurescores = pd.concat([dfcolumns, dfscores], axis = 1)
featurescores.columns = ["feature_name", "feature_score"]
print(featurescores.nlargest(2, "feature_score"))
```
